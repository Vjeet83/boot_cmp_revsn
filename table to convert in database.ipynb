```python
# faculty table 
import pandas as pd
import random

# List of 30 random Hindu names suitable for adults or teachers
random_hindu_names = [
    "Mohan", "Kamala", "Arjun", "Kavita", "Rajesh", "Priya", "Sanjay", "Meena", "Ravi", "Anjali",
    "Manoj", "Sunita", "Deepak", "Nisha", "Rahul", "Pooja", "Ajay", "Alka", "Vikram", "Reena",
    "Ashok", "Seema", "Vinay", "Radhika", "Prakash", "Shilpa", "Karan", "Sangeeta", "Sachin", "Geeta"
]

# Sample real addresses and departments
sample_addresses = [
    f"{random.randint(100, 500)}- {random.choice(['Jaipur', 'Jodhpur', 'Bundi', 'Kolkata', 'Kota', 'Antela', 'Sikar', 'Avara'])}"
    for _ in range(30)
]

sample_departments = [
    "Computer Science", "Mathematics", "Physics", "Chemistry", "Biology"
]

# Generate data for the faculty table
faculty_data = {
    'faculty_id': [i+1 for i in range(30)],  # Generate faculty IDs
    'faculty_name': random.sample(random_hindu_names, 30),  # Randomly choose 30 names from the random_hindu_names list
    'age': [random.randint(30, 60) for _ in range(30)],  # Generate random ages
    'gender': ['Male' if random.random() > 0.5 else 'Female' for _ in range(30)],  # Randomly assign gender
    'contact_number': ['+123456789' + str(random.randint(1000, 9999)) for _ in range(30)],  # Generate contact numbers
    'address': random.sample(sample_addresses, 30),  # Sample realistic addresses
    'department': random.choices(sample_departments, k=30)  # Sample realistic department names
}

# Create DataFrame for faculty table
df_faculty = pd.DataFrame(faculty_data)

# Display the DataFrame
# print(df_faculty)
# print(df_faculty.shape)
df_faculty
```

```python
# faculty attendance
random.seed(544)
import pandas as pd
import random

faculty_ids = [i+1 for i in range(30)]  
# Generating dates for 9 days 
date_range = pd.date_range('2023-12-01', periods=9, freq='D')
date_columns = [date.strftime('%Y-%m-%d') for date in date_range]

# Creating dictionary to store attendance data
faccult_attendance_data = {
    'faculty_id': faculty_ids
}

# Adding columns for 5 days with random attendance statuses
for date in date_columns:
    faccult_attendance_data[date] = random.choices(['present', 'presen', 'present','absent','present', 'leave','absent'], k=len(faculty_ids))

# Create DataFrame for attendance table
df_faculty_attendance = pd.DataFrame(faccult_attendance_data)
path="C://Desktop//faculty_attendence.csv"
# Display the DataFrame
# print(df_faculty_attendance)
df_faculty_attendance.to_csv(path,index=False)
```
